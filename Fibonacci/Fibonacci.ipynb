# Fibonacci

Primero se hizo el siguiente código para comprobar cuánto se demoraba el programa en llegar al valor de la posición dada por el usuario, y en este caso no se da overflow, o por lo menos no en los casos con los que se probó, ya que los valores de python que no utilizan numpy tienen un tamaño arbitrario que depende sólo de la memoria: 

```python
import time

start_time = time.time()
cant=int(input())
n1=1;
n2=1; 
for i in range(2,cant):
    
   n2 = n1 + n2;
   n1 = n2 - n1;
    
print("Tiempo: %s segundos." % (time.time() - start_time))
```

Después, con un ciclo para comparar el tiempo con el número de valores n que se piden:

```python
import time
import numpy as np
import matplotlib.pyplot as pl
from __future__ import unicode_literals
%matplotlib inline

N=int(input())
arr=np.arange(0.000,N,dtype=np.float_)

for n in range(N):
    
    start_time = time.time()
    
    cant=n;
    n1=1;
    n2=1; 
    
    for i in range(2,cant):
    
        n2 = n1 + n2;
        n1 = n2 - n1;
        
    tfinal=time.time() - start_time
    
    arr[n]= float(tfinal)
    
pl.ylabel("Tiempo")
pl.xlabel('Tamaño')
pl.plot(arr)
```
